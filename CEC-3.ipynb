```python
# SET - 1
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Load dataset
df = pd.read_csv("student_performance.csv")

# Display first and last 8 records
print(df.head(8))
print(df.tail(8))

# Fill missing values in Marks
df['Marks'].fillna(30, inplace=True)

# Box plot for Attendance
plt.boxplot(df['Attendance'])
plt.title("Attendance Box Plot")
plt.show()

# Section B

# Convert Result to numeric
df['Result'] = df['Result'].map({'Fail':0, 'Pass':1})

X = df[['StudyHours','Attendance','Marks']]
y = df['Result']

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

# Apply KNN
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)

# Prediction
y_pred = model.predict(X_test)

# Accuracy & Confusion Matrix
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Custom prediction
print("Prediction:", model.predict([[6,80,70]]))
```

```python
# SET - 2
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Load dataset
df = pd.read_csv("employee_performance.csv")

# Display last 5 records
print(df.tail())

# Summary
print(df.info())
print(df.describe())

# Bar chart (Average Salary by Performance)
df.groupby('Performance')['Salary'].mean().plot(kind='bar')
plt.title("Average Salary by Performance")
plt.show()

# Section B

# Convert target to numeric
df['Performance'] = df['Performance'].map({'Low':0,'Medium':1,'High':2})

X = df[['Age','Experience','Salary']]
y = df['Performance']

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

# Decision Tree
model = DecisionTreeClassifier(max_depth=4)
model.fit(X_train, y_train)

# Prediction
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Custom prediction
print("Prediction:", model.predict([[35,8,60000]]))
```

```python
# SET - 3
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# Load dataset
df = pd.read_csv("patient_diagnosis.csv")

# Display first 8 records
print(df.head(8))

# Shape and columns
print("Shape:", df.shape)
print("Columns:", df.columns)

# Scatter plot
plt.scatter(df['GlucoseLevel'], df['BMI'])
plt.xlabel("GlucoseLevel")
plt.ylabel("BMI")
plt.title("Glucose vs BMI")
plt.show()

# Section B

# Convert target
df['Disease'] = df['Disease'].map({'No':0,'Yes':1})

X = df[['Age','BloodPressure','GlucoseLevel','BMI']]
y = df['Disease']

# Split (30% test)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)

# Logistic Regression
model = LogisticRegression()
model.fit(X_train, y_train)

# Prediction
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Custom prediction
print("Prediction:", model.predict([[45,130,150,28]]))
```

```python
# SET - 4
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix

# Load dataset
df = pd.read_csv("customer_purchase.csv")

# Shape and columns
print(df.shape)
print(df.columns)

# Histogram
plt.hist(df['Salary'])
plt.title("Salary Distribution")
plt.show()

# Section B

# Convert target
df['Purchase'] = df['Purchase'].map({'No':0,'Yes':1})

X = df[['Age','Salary']]
y = df['Purchase']

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

# SVM
model = SVC(kernel='linear')
model.fit(X_train, y_train)

# Prediction
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Custom prediction
print("Prediction:", model.predict([[30,50000]]))
```

```python
# SET - 5
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# Load dataset
df = pd.read_csv("loan_prediction.csv")

# Display last 4 rows
print(df.tail(4))

# Check missing values
print(df.isnull().sum())

# Pie chart
df['Income'].value_counts().plot(kind='pie')
plt.title("Income Distribution")
plt.show()

# Section B

# Convert target
df['Approved'] = df['Approved'].map({'No':0,'Yes':1})

X = df[['Income','CreditScore','LoanAmount']]
y = df['Approved']

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

# Logistic Regression
model = LogisticRegression()
model.fit(X_train, y_train)

# Prediction
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Custom prediction
print("Prediction:", model.predict([[50000,700,20000]]))
```
